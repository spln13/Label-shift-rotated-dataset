```python
import matplotlib.pyplot as plt
import numpy as np
from data_loader import load_data_for_client

def plot_client_images(data_loader, client_id, num_images=10):
    fig, axes = plt.subplots(1, num_images, figsize=(15, 1.5))
    plt.suptitle(f'Client {client_id} - First {num_images} images', fontsize=13, y=1.1)
    
    for images, labels in data_loader:
        for i in range(num_images):
            ax = axes[i]
            img = images[i].numpy()
            img = np.transpose(img, (1, 2, 0))
            img = img * 0.5 + 0.5  # 反归一化，假设原始归一化是以0.5为中心
            ax.imshow(img)
            ax.set_title(f'Label: {labels[i]}')
            ax.axis('off')
        break
    plt.show()


num_clients = 20

for client_id in range(num_clients):
    client_data_loader = load_data_for_client(client_id, dataset='cifar10', dataset_type='train')
    plot_client_images(client_data_loader, client_id)

for client_id in range(num_clients):
    client_data_loader = load_data_for_client(client_id, dataset='mnist', dataset_type='train')
    plot_client_images(client_data_loader, client_id)
```
